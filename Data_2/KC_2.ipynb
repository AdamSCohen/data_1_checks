Reading in the data and cheking the shape.
The data is a list of spells from Dungeons & Dragons.

```python
r = requests.get("https://api.open5e.com/spells/?limit=1000")
r2 = r.json()['results']
df = pd.DataFrame(data=r2)
df.head()
```

```python
df.count()
```

I noticed that I returned more entires than I expected. The spells are coming from a viriety of sources, but I only want the ones from the Players Handbook, refered to in this data as "Systems Reference Document".
I then check for all documents used in the data.

```python
docs = df["document__title"].unique()
print(docs)
```

I remove all rows with other document titles other than "Systems Reference Document". There are now 319 rows, which is what I expected initially.

```python
df = df.loc[df['document__title'] == 'Systems Reference Document']
df.head()
```

```python
df.count()
```

Remove columns that I will not need for the final project.

```python
df.drop(columns=['page', 'level', 'material', 'archetype', 'circles', 'document__slug', 'document__title', 'document__license_url'],
        inplace=True)
df.columns
```

Change the name of some columns to better match other data that I plan to use in my final project.

```python
df.rename(columns={'slug' : 'key', 'level_int' : 'level', 'dnd_class' : 'class'}, inplace=True)
df.columns
```
